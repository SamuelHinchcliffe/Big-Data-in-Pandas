Splits Bigdata into chunks and runs a function over each chunk returning the new data over multiple chunked Data Frames after which we combine the Dataframes back together to create a single cleaner/smaller Data Frame. 

```python
import pandas as pd
# Specifies Chunksize & Chunk_no
chunk_size = 10000000
chunk_no = 1
# Runs a for loop itterating a function over each chunk of the DB
for chunk in pd.read_csv('bigdata.csv', chunksize = chunk_size):
# Specifies column names for each chunk
    chunk.columns = ['column 1', 'column 2' , 'column 3', 'column 4', 'column 5', 'column 6', 'column 7', 'column 8']
    class Funct:
        def __init__(self, func):
            self.func = func   
# Filtering function to run over each chunk once called upon.
    new_chunk = Funct(chunk.loc[chunk['column 8'] > 30000]) 
# Creates a new variable calling the function as functions cannot be exported as a file
    super_chunk = new_chunk.func
# Exports the chunks specifying file location/name + the chunk number as a str to allow for the creation of multiple files + .csv 
    super_chunk.to_csv('Chunks\chunk'+str(chunk_no)+'.csv',index=False)
    chunk_no += 1

```

```python
# This rejoins the chunked files into a clean data frame without overlaping data or headers
import glob
import os

# Parses in your pathway specifying as a rawstring
path =  r'Pathway'
# Variable with set parameter to specify all files containing .csv 
all_chunks = glob.glob(os.path.join(path, "*.csv"))
# Reads all files specified by the all_chunks varible
each_chunk = (pd.read_csv(c) for c in all_chunks)
#Concatenates all of the df's together and returns a single shrunken df containing only relevant info
output = pd.concat(each_chunk, ignore_index=True)
output.to_csv('output.csv', index = False )
```
